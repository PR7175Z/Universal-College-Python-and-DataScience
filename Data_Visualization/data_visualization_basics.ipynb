<h3>Plot </h3>

```python
data_frame_new_1=pd.DataFrame({
    'Price': [10,20,30,40,50,60,70],
    'Sold Units': [20,40,60,80,100,120,140]
})


data_frame_new_2=pd.DataFrame({
    'Price': [10,20,30,40,50,60,70],
    'Sold Units': reversed([20,40,60,80,100,120,140])
})

display(data_frame_new_1)
plt.plot(data_frame_new_1['Price'],data_frame_new_1['Sold Units'])
plt.plot(data_frame_new_2['Price'],data_frame_new_2['Sold Units'])
plt.legend(['Data_Frame_1','Data_Frame_2'])
# plt.scatter(x=data_frame_new_1['Price'],y=data_frame_new_1['Sold Units'])
plt.xlabel('Price')
plt.ylabel('Sold units')
```

<h3> Plotting Line dyanamically </h3>

```python
import random

def generate_y(x):
    m,c=-5,0
    
    return m*x+c

x=[random.randint(1,10) for i in range(1,10)]
y=[generate_y(i) for i in x]

plt.plot(x,y)
```

<h3> Create Parabola dynamically symmetric about y axis</h3>

```python
import random
list_val=[random.randint(1,10) for i in range(1,random.randint(10,50))]
counter={}

for i in list_val:
    counter[str(i)]=list_val.count(i)


print(counter)
plt.bar(x=counter.keys(),height=counter.values())
plt.xlabel('numbers')
plt.ylabel('count')

```
